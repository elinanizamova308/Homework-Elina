```python
import numpy as np
import matplotlib.pyplot as plt

a = np.loadtxt('http://www.scipy-lectures.org/_downloads/populations.txt') 
print(a)
Years = a[:,0]
H = a[:,1]
L = a[:,2]
C = a[:,3]
```

```python
#1 Plot the population size of each species for each year on a single graph. Add legend and axis labels to the plot.
plt.plot(Years, H, label='Hares')
plt.plot(Years, L, label='Lynxes', c='magenta')
plt.plot(Years, C, label='Carrots', c='aqua')
plt.xlabel('Year', fontsize=15)
plt.ylabel('Quantity', fontsize=15)
plt.xticks(np.arange(1900, 1921, step=1),fontsize=6 )
plt.grid()
plt.legend()
plt.show()
```

```python
#2 Find which year each species had the largest population. Mark these years in the plot of populations.

#MaxPopInd = np.argmax(inf, axis = 0)[1:]
plt.scatter(Years[np.argmax(H)], np.max(H))
plt.scatter(Years[np.argmax(L)], np.max(L))
plt.scatter(Years[np.argmax(C)], np.max(C))
plt.plot(Years, H, label='Hares')
plt.plot(Years, L, label='Lynxes', c='magenta')
plt.plot(Years, C, label='Carrots', c='aqua')
plt.xlabel('Year', fontsize=15)
plt.ylabel('Quantity', fontsize=15)
plt.xticks(np.arange(1900, 1921, step=1),fontsize=6 )
plt.grid()
plt.legend()
plt.show()
```

```python
#3 Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for each year in the period.

fig, ax = plt.subplots(figsize=(15, 10))
Hnew = H/(H+L)*100
Lnew = L/(H+L)*100
ax.bar(Years, Hnew,  label = 'Hares')
ax.bar(Years, Lnew, bottom = Hnew, label='Lynxes')
ax.set_ylabel('Percents', fontsize=15)
ax.set_xlabel('Year', fontsize=15)
ax.legend()
plt.xticks(np.arange(1900, 1921, step=1))
plt.show()
```

```python
#4 Find the mean and std of the populations of each species. 
#Plot the historgam of population for each species and show mean values with vertical line. 
#Show each histogram in a separate plot. Arrange 3 plots in a row.

fig = plt.figure(figsize = (15,15))

print("Hares mean:", np.mean(H), "; Hares std:", np.std(H))
print("Lynxes mean:", np.mean(L), "; Lynxes std:", np.std(L))
print("Carrots mean:", np.mean(C), "; Carrots std:", np.std(C))
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 7))

ax1.hist(H, color = 'goldenrod')
ax1.set_title('Hares')
ax1.axvline(x=np.mean(H), c = 'black')
ax2.hist(L, color = 'turquoise')
ax2.set_title('Lynxes')
ax2.axvline(x=np.mean(L), c = 'black')
ax3.hist(C, color = 'indigo')
ax3.set_title('Carrots')
ax3.axvline(x=np.mean(C), c = 'black')

plt.show()
```

```python
#9 Написать функцию, реализующую алгоритм kNN.
from sklearn.datasets import fetch_openml

#Загрузка может занять около минуты
images, labels = fetch_openml('mnist_784', return_X_y=True, as_frame=False)

i = 125
plt.imshow(images[i].reshape(28, 28), cmap='Greys_r')
plt.title('Number: ' + labels[i])
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(images, labels, test_size=0.3)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
def kNN(train_X, train_Y ,test_X, k):
    res = []
    test_Y = []
    for i in test_X:
        #print(np.linalg.norm(train_X - i, axis=1))
        res = np.argsort(np.linalg.norm(train_X - i, axis=1))[:k]   # в количестве k сортируем индексы по расстоянию от матрицы обучающего до матрицы тестового
        numberOftrain = np.argmax(np.unique(train_Y[res],  return_counts = True)[1])   # количество групп объектов, которым принадлежат эти k элементов
        test_Y.append(np.unique(train_Y[res],  return_counts = True)[0][numberOftrain]) # та группа, которой принадлежит i
    return np. array(test_Y)

def accur(test_X, test_Y):
  accur = 0
  l = len(test_X)
  for i in range(l):
    if (test_X[i] == test_Y[i]):
      accur +=100/l
  return (accur)


l=[]
for i in range(1, 101):
  l.append(accur(kNN(x_train[:1000],y_train[:1000],x_test[:100],i), y_test[:100]))
print ("max accur:", np.max(l), "max k:", np.argmax(l)+1 )
```
